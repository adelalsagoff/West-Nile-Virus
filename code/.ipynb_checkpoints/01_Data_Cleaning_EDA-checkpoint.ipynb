# Introduction and  Problem Statement

West Nile virus (WNV) is the leading cause of mosquito-borne disease in the continental United States *([Source](https://www.cdc.gov/westnile/index.html))*. Every week from late spring through the fall, mosquitos in traps across the city are tested for the virus. The results of these tests influence when and where the city will spray airborne pesticides to control adult mosquito populations *([Source](https://www.kaggle.com/c/predict-west-nile-virus))*.
 
 Although pesticides are known to be effective in dealing with the virus-carrying mosquitoes, it is expensive to deploy pesticides throughout the city of Chicago. As data scientists, we want to understand the factors driving the spread of WNV by leveraging on data collected by Chicago's weather stations. We also want to develop a classfication model that could predict the presence of WNV within the area of Windy City. Through these studies, we hope to suggest a cost-efficient and effective method of deploying pesticides within the area.

## Check on Data
A function is defined to have a quick check on the Data Type, Rows, Columns, Null Rows and Duplicate Rows

```python
def basic_info(df):
    ''' Input Dataframe
    Output Name of Column & its Data Type, Number of Rows/Columns, Number of Null/Duplicate Rows'''
    print("========== Summary ==========")
    print(df.info())
    
    print("\n=========== Shape ===========")
    print(f"(row, column) = {df.shape}") # prints the shape of the dataset
    
    print("\n====== Total Null/Dup =======")
    null_row = df.isnull().sum().sum()
    print(f"Total Null Rows: {null_row} ({round(null_row*100/len(df),2)}%)")
    dup_row = df[df.duplicated(keep=False)].shape[0]
    print(f"Total Duplicate Rows: {dup_row} ({round(dup_row*100/len(df),2)}%)")
     
    if null_row > 0:
        print("\n========= Null Rows =========")
        print(df.isnull().sum())
        
        sns.set(); #setting seaborn as default background
        msno.bar(df)
        plt.title("Missing value count for dataset",fontsize = 30)
    
```

#### Identifying missing data and impute as np.nan

Based on the weather documentation provided with the dataset, missing data are repesented as 'M'. 

Values not available for sunrise/sunset are represented as '-'.

Additional trace data, defined as precipitation, such as rain or snow, that is greater than zero, but is too small to be measured by standard units or methods of measurement, is represented as 'T'.

```python
#show counts of unique value in station
weather.Station.value_counts()
```

```python
#show counts of unique value in depart
weather.Depart.value_counts()
```

```python
#show counts of unique value in snowfall
weather.SnowFall.value_counts()
```

<b>comment</b>:

Missing values for 'Depart' column and Snowfall are the same number. This corresponds to rows where 'Station' value is 2.

This is inline with the weather documentation that not all station 2 reports snowfall totals.

Furhter above value counts of Snowfall also shows trace indicated as '  T' as well.

Therefore the following values, 'T', '  T', '-', 'M' and ''(just in case) will be replaced as np.nan.

```python
# create applymap function to apply to all column in dataset

def replace_miss_val(x):
    if str(x) in ['T',' T', '-', 'M', '','', '  T']: #check if str value is in list
        return np.nan #if str is in list, replace to nan value
    else:
        return x
```

```python
# apply function, show first five rows
weather = weather.applymap(replace_miss_val)
weather.head()
```

```python
# convert columns that should be numeric into numeric values
weather.Tavg = pd.to_numeric(weather.Tavg)
weather.Heat = pd.to_numeric(weather.Heat)
weather.Cool = pd.to_numeric(weather.Cool)
weather.PrecipTotal = pd.to_numeric(weather.PrecipTotal)
weather.StnPressure = pd.to_numeric(weather.StnPressure)
weather.SeaLevel = pd.to_numeric(weather.SeaLevel)
weather.AvgSpeed = pd.to_numeric(weather.AvgSpeed)
weather.SnowFall = pd.to_numeric(weather.SnowFall)
weather.Depth = pd.to_numeric(weather.Depth)
weather.WetBulb = pd.to_numeric(weather.WetBulb)
```

From the above WnvPresent graph, we can observe that occurences of WNV peaks around week 33-34 (around August). In Chicago, August is during the summer time and therfore is very warm. Temperature usually peaks around July (above we have a plot of mean temperature by WeekofYear).

As we can see from the temperature curve, the weather starts getting warm after week 23 and peaks between week 28 to 35, before gradually cooling off. Dewpoint also follows this pattern of increasing after week 23 before decreasing from week 35.

Additionally, we can see that during the period of week 25 to 35, mean Dewpoint is above 50, therefore suggesting high amount of moisture in the air. 

This coincides with our subject background research as well.

Culex mosquitos favor high humidity conditions as they function best in these environments. They usually require around 2 weeks to grow from larvae to an adult, as we see cases slowly rising at week 28. There will be a very steep rise in cases up till the peak around week 33/34. Which coincides with the period of week 25 to 35, where average temperature hovers around the high of 72.5 to 76 degrees.

Just as temperature drops after week 30, and Dewpoint after week 34, the presence of WNV in the traps tend to fall sharply at week 35 onwards.

## EDA: Heatmap of Traps and WNVPresent in Respective Years

```python
def plot_density_map(data, ax, show_traps=True):
    # Filter by year of interest
    infections = data[data['WnvPresent'] == True]
    # Get df where each row represents a day where a trap tested positive for WNV at least once.
    daily_infections = infections.groupby(['Trap', 'Date', 'Latitude', 'Longitude']).max()['WnvPresent'].reset_index()
    daily_infections_ll = daily_infections[['Longitude', 'Latitude']]

    # Estimate density
    kd = KernelDensity(bandwidth=0.02)
    kd.fit(daily_infections_ll.values)
    lon_min, lon_max = -88, -87.5
    lat_min, lat_max = 41.6, 42.1
    n = 100
    xx,yy = np.linspace(lon_min, lon_max, n), np.linspace(lat_min, lat_max, n)
    xx, yy = np.meshgrid(xx, yy)
    X = np.vstack([xx.ravel(), yy.ravel()]).T
    zz = np.exp(kd.score_samples(X)).reshape(n, n)

    # Plot the map and density contours
    mapdata = np.loadtxt("../assets/mapdata_copyright_openstreetmap_contributors.txt")
    traps = train_df[['Date', 'Trap','Longitude', 'Latitude', 'WnvPresent']]
    lon_lat_box = (lon_min, lon_max, lat_min, lat_max)
    aspect = mapdata.shape[0] * 1.0 / mapdata.shape[1]
    cnt = ax.contourf(xx,yy,zz, cmap='coolwarm', alpha=0.5)
    plt.colorbar(cnt, ax=[ax])
    ax.imshow(mapdata, 
           cmap=plt.get_cmap('gray'), 
           extent=lon_lat_box, 
           aspect=aspect)
    
    # plot trap locations
    if show_traps:
        trap_locations = traps[['Longitude', 'Latitude']].drop_duplicates().values
        ax.scatter(trap_locations[:,0], trap_locations[:,1], marker='x', c='black')
    
def gen_plot_data(data, year):
    return data[data['Date'].dt.year == year]
    
gs = gridspec.GridSpec(2, 2)
fig = plt.figure(figsize=(18,18))
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

plot_density_map(gen_plot_data(train_df, 2007), ax1)
plot_density_map(gen_plot_data(train_df, 2009), ax2)
plot_density_map(gen_plot_data(train_df, 2011), ax3)
plot_density_map(gen_plot_data(train_df, 2013), ax4)

ax1.set_title(2007, fontsize=16)
ax2.set_title(2009, fontsize=16)
ax3.set_title(2011, fontsize=16)
ax4.set_title(2013, fontsize=16);
plt.savefig("../images/traps_wnvpresent_heatmap.png",bbox_inches = 'tight');
```

```python
spray_locations_11 = spray[spray['Date'].dt.year == 2011]
spray_locations_13 = spray[spray['Date'].dt.year == 2013]

def plot_spray_map(data, ax, show_spray_year):
    # Filter by year of interest
    infections = data[data['WnvPresent'] == True]
    #Get df where each row represents a day where a trap tested positive for WNV at least once.
    daily_infections = infections.groupby(['Trap', 'Date', 'Latitude', 'Longitude']).max()['WnvPresent'].reset_index()
    daily_infections_ll = daily_infections[['Longitude', 'Latitude']]

    # Estimate density
    kd = KernelDensity(bandwidth=0.02)
    kd.fit(daily_infections_ll.values)
    lon_min, lon_max = -88, -87.5
    lat_min, lat_max = 41.6, 42.1
    n = 100
    xx,yy = np.linspace(lon_min, lon_max, n), np.linspace(lat_min, lat_max, n)
    xx, yy = np.meshgrid(xx, yy)
    X = np.vstack([xx.ravel(), yy.ravel()]).T
    zz = np.exp(kd.score_samples(X)).reshape(n, n)

    # Plot the map and and density contours
    mapdata = np.loadtxt("../assets/mapdata_copyright_openstreetmap_contributors.txt")
    spray = pd.read_csv('../assets/spray.csv')[['Date','Longitude', 'Latitude']]
    lon_lat_box = (lon_min, lon_max, lat_min, lat_max)
    aspect = mapdata.shape[0] * 1.0 / mapdata.shape[1]
    cnt = ax.contourf(xx,yy,zz, cmap='coolwarm', alpha=0.5)
    plt.colorbar(cnt, ax=[ax],shrink=0.50)
    ax.imshow( mapdata, 
           cmap=plt.get_cmap('gray'), 
           extent=lon_lat_box, 
           aspect=aspect)
    
    # plot spray locations

    if show_spray_year == 2011:
        #spray_locations = spray[spray['Date'].dt.year == 2011]
        spray_locations = spray_locations_11[['Longitude', 'Latitude']]
        spray_locations = spray_locations[spray_locations['Latitude'] < lat_max].values
        ax.scatter(spray_locations[:,0], spray_locations[:,1], marker='o', c='black',alpha = 0.006)
    if show_spray_year == 2013:
        #spray_locations = spray[spray['Date'].dt.year == 2011]
        spray_locations = spray_locations_13[['Longitude', 'Latitude']]
        spray_locations = spray_locations[spray_locations['Latitude'] < lat_max].values
        ax.scatter(spray_locations[:,0], spray_locations[:,1], marker='o', c='black',alpha = 0.006)
          
        
    
def gen_plot_data(data, year):
    years = data['Date'].map(lambda x: x.year)
    mask = years == year
    return data[mask]
    
gs = gridspec.GridSpec(1, 2)
fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])


plot_spray_map(gen_plot_data(train_df, 2011), ax1, 2011)
plot_spray_map(gen_plot_data(train_df, 2013), ax2, 2013)

ax1.set_title(2011, fontsize=16)
ax2.set_title(2013, fontsize=16)
plt.savefig("../images/spray_location_heatmap.png",bbox_inches = 'tight');
```

## Cost Benefit Analysis

#### Febrille Ilness and Serious Symptoms when infected with WNV

According to the Centers of Disease Control and Prevention (CDC) *([Source](https://www.cdc.gov/westnile/symptoms/index.html))*, there are two main outcomes that the team believes would result in economic cost. 

Febrile illness (WNF):
- About 1 in 5 people who are infected develop a fever along withh other symptoms such as headache, body aches.
- Most people with this type of West Nile virus disease recover completely, but feeling of fatigue can last for weeks or months.
    - We will assume that infected people are working adults that would require 2 weeks on average to recover, and would therefore miss work during this period. 

West Nile neuroinvasive disease (WNND):
- About 1 in 150 people who are infected develop a severe illness affecting the central nervous system such as encephalitis (inflammation of the brain) or meningitis (inflammation of the membranes that surround the brain and spinal cord).
- This would require the person to stays in a hospital while receiving medical care or treatment or in some cases receive treatments as outpatients.
- We will assume these patients are unable to work for 60 days.

#### Using Sacremento County as an Example *([Source](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3322011/#:~:text=The%20total%20economic%20impact%20of%20the%202005%20WNV%20disease%20outbreak,the%20emergency%20spray%20is%20uncertain.))*

Sacremento was decided as a proxy as comprehensive research had already been done on the cost associated with contracting WNV.

In 2005, an outbreak of WNV disease occurred in Sacramento County, California where 163 human cases were reported. In response, Sacramento-Yolo Mosquito and Vector Control District conducted an emergency aerial spray.

The economic impact of the outbreak had cost the county an approximate 2.28 million dollars in medical treatment and patients productivity loss for both West Nile fever and West Nile neuroinvasive disease. Vector control cost an approximate 701,790 dollars , including spray procedures and overtime hours. The total economic impact of WNV was approximately 2.98 million dollars.

From this study's cost-benefit analysis indicated that only 15 cases of West Nile neuroinvasive disease would need to be prevented to make the emergency spray cost-effective.

##### Average Costs in terms of medical and productivity loss

| Type of Cost                                 | WNF Patient | WNND Patient |
|----------------------------------|-------------|--------------|
| Medical                          | \$302        | \$33,143      |
| Productivity Loss (per work day) | \$191        | \$191         |
| Productivity Loss (per non work day)  | \$125        | \$125         |

The numbers above are an approximation of the average cost per patient type in Sacremento as with each patient, the extent of treatment recieved would vary. We will use this as a proxy for the case of Chicago.

##### Cost to Spray in 2011 in Chicago
Seeing from the image below, the area where spray activity had occured in 2011 was at O'Hare internation airport, an area of approximately 23.33 km2. The area which Sacremento had calculated the cost to spray for an area of 477 km2, was 1,438,619 dollars. Therefore, giving us an estimate cost of spray in Chicago of 70,362 dollars for an aera of 23.33 km2.

##### Estimating Productivity Loss Cost in 2011 in Chicago

Assuming 150 people had tested positive to WNV in 2011. If 1 in 5 people develop WNF symptoms and 1 in 150 people develop WNND symptoms, 20 people with WNV would result in economic cost of approximately 81,360 dollars ((\\$302 + \\$191\*10 work days + \\$125\*4 non work days) \*30 patients), while 1 person with WNND would result in economic cost of approximately 43,547 dollars (\\$33,143 + \\$191\*44 work days + \\$125\*16 non work days).

This would result in an estimate of 124,907 dollars in total economic cost in 2011.

```python
def df_creation(columns):
    """
    Function to create a new collated dataframe for weather dataset. 
    input: numeric columns to calculate the averages between station1 and station2
    outputs: the averages of the columns
    """
    for i in range(len(columns)):
        final_weather[columns[i]] = (weather.loc[weather['Station']==1][[columns[i]]].reset_index(drop=True)+weather.loc[weather['Station']==2][[columns[i]]].reset_index(drop=True)/2)
    return final_weather.head()
```

#### Sunrise and Setset Time

##### Features that Desribe the Sunrise and Setset Time

We have decided to convert the sunrise/sunset time to seconds as we believed that machines do not process time similarly as us. For instance, we read 1840 as 18 hours and 40 minutes but the machine only reads 1840 as an integer. That being said, we understand that there are a maximum of 24 hours and 60 minutes but not for machine. Therefore, we decided to scale the timing to seconds as there can only be 86400 seconds a day and it is more intuitive for both machine and us. 



```python
# Convert sunrise & sunset into seconds from start of day
def get_sec(time_str):
    """
    inputs the time
    outputs the seconds
    """
    hours = int(time_str)//100 # Get the hour value
    mins = int(time_str) % 100 # Get the minute value
    total_seconds = hours * 3600 + mins * 60
    return total_seconds
```

```python
final_weather.Sunset = final_weather.Sunset.apply(get_sec)
```

```python
final_weather.Sunrise = final_weather.Sunrise.apply(get_sec)
```
