<center><h3>Problem Statement</center></h3>
<br>

According to Millionacres' Home Buyer & Seller Survey, 52% of homeowners have concerns about selling their homes pedominantly due to high uncertainty with regards to property valuation. As a data scientist for a property technology company, a model is to be built to conduct higher accuracy valuation and optimise the price listing found on the company's real estate portal.

This model will be trained on historical transactions in Ames property market, and in the process of building this model, synthesize and highlight the best features that brings the most value to their home and provide company's clientle to make more informed decisions.

<center><h3>Data Cleaning</center></h3>
<br>

The Ames dataset began with with 79 features. In order to avoid problems associated with high dimesionaltity, the number of features used to input into our model should be reduced.

Features with missing values percentage greater than 80 percent were removed from the dataset. As well as features that had frequency of the mode value greater than 80 percent as they provide little information and their coefficient would affect our model. A heatmap was used to visualise missing values seen in Figure 1 and 2 for numerical and categorical features respectively.

<center><h3>Conclusion</center></h3>

The ridge model performs fairly well for house price up to 400,000 dollars within 100,000 dollars boundary, however predicted two outliers which overvalues 300,000 dollars to 400,000 dollars above the actual sale price. This however may not be the most optimal model to present to the board of directors as other models could be used to provide more accurate predictions. 

<br>
<center><h3>Recommendations</center></h3>

As the model increase in bias when predicting house prices above 400,000 dollars what could be done is gather more data on house that cost higher than 400,000 dollars.

Additionally, handpicking specific features that have direct correlation to improving house prices, and informing the campanies customer base of these features can be a value added service. Customers would also benefit from making more informed decisions to increase the sale price of their homes. 

A hypothesis test could also be done to test if those feature are indeed correlated to house prices.
